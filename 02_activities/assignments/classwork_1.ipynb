```python

np.random.seed(613)

x = np.arange(50)
y = np.random.randint(0,100,50)

# Creating a figure and an axes
fig, ax = plt.subplots (figsize = (5,3))

ax.scatter(x,y)
```

```python
fig, ax = plt.subplots(figsize=(5, 3))

# Create a histogram using y data
ax.plot(x,y)

# Add titles and labels
ax.set_title('Total growth over time')
ax.set_xlabel('Years since start')
ax.set_ylabel('Total growth')
fig.tight_layout
```

```python
font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'monospace','color':'green','size':14}

fig, ax = plt.subplots(figsize=(5, 3))

ax.plot(x, y)

# Add titles and labels
ax.set_title('Total growth over time', fontdict=font1, loc = 'right') 
ax.set_xlabel('Years since start', fontdict=font2)
ax.set_ylabel('Total growth', fontdict=font2, loc = 'top')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Set seed for reproducibility
np.random.seed(613)

# create data
values = np.random.uniform(size=40)

(markers, stemlines, baseline) = plt.stem(values)
plt.setp(markers, marker='D', markersize=10, markeredgecolor="orange", markeredgewidth=2, color='red')
plt.setp(stemlines, color='gray', linewidth=1)
plt.setp(baseline, color='black', linewidth=1)

# Customize titles and labels
plt.title('Total growth over time', fontsize=16, color='blue')
plt.xlabel('Years since start', fontsize=12, color='green')
plt.ylabel('Total growth', fontsize=12, color='green')
```
